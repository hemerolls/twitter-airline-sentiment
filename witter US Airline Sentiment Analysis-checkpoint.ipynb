```python
# Заполним пропуски
df['negativereason'] = df['negativereason'].fillna('Unknown')

# Label Encoding
label_encoders = {}
categorical_features = ['airline', 'negativereason']

for col in categorical_features:
    le = LabelEncoder()
    df[col + '_encoded'] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Проверим, установлены ли ресурсы
try:
    stop_words = set(stopwords.words('english'))
except LookupError:
    print("Загружаем ресурсы NLTK...")
    nltk.download('stopwords')
    nltk.download('punkt')
    stop_words = set(stopwords.words('english'))

def clean_text(text):
    """
    Очистка и предобработка текста
    """
    if not isinstance(text, str):
        return ""
    
    # Удаление ссылок, знаков и пр.
    text = re.sub(r'http\S+', '', text)      # удаление ссылок
    text = re.sub(r'@\w+', '', text)         # удаление @username
    text = re.sub(r'#\w+', '', text)         # удаление хэштегов
    text = re.sub(r'[^A-Za-z\s]', '', text)  # оставляем только буквы
    
    try:
        tokens = word_tokenize(text.lower())
        tokens = [w for w in tokens if w not in stop_words and len(w) > 2]
        return ' '.join(tokens)
    except Exception as e:
        print(f"Ошибка при обработке текста: {e}")
        return ""

# Применяем функцию
df['cleaned_text'] = df['text'].apply(clean_text)
```

```python
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_text = vectorizer.fit_transform(df['cleaned_text'])
```

```python
from scipy.sparse import hstack

# Категориальные признаки
X_cat = df[['airline_encoded', 'negativereason_encoded']].values

# Объединяем текстовые и категориальные признаки
X_combined = hstack([X_text, X_cat])
y = df['airline_sentiment']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=model.classes_, yticklabels=model.classes_)
plt.title("Confusion Matrix")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
